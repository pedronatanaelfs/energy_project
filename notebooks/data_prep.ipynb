```python
# Importing Libraries
import pandas as pd
```

```python
# Reading files
base_path = r'C:\Users\Pedro Natanael\OneDrive\Documents\Mestrado\Reconhecimento de Padrões\energy_project\data'

year_solo = [f'_{year}' for year in range(2006, 2022)]

years_month_full = [f'-2_{year}_{str(month).zfill(2)}' for year in [2022, 2023] for month in range(1, 13)]

final_year = [f'-2_{2024}_{str(month).zfill(2)}' for month in range(1, 7)]

years = year_solo + years_month_full + final_year

dataframes = {}

for year in years:
    file_path = f'{base_path}\GERACAO_USINA{year}.csv'
    data = pd.read_csv(file_path, sep=';')

    dataframes[str(year)] = data

dataframes['_2006'].head()
```

```python
for key in dataframes:
    dataframes[key] = dataframes[key].query("nom_tipousina == 'EOLIELÉTRICA'")
```

```python
for key in dataframes:
    unique_values = dataframes[key]['nom_tipousina'].unique()
    print(f"Unique values in 'nom_tipousina' for {key}: {unique_values}")
```

```python
for key in dataframes:
    dataframes[key] = dataframes[key][['din_instante', 'val_geracao']]
```

```python
for key in dataframes:
    print(dataframes[key].dtypes)
```

```python
for key in dataframes:
    dataframes[key]['din_instante'] = pd.to_datetime(dataframes[key]['din_instante'])
    print(dataframes[key].dtypes)
```

```python
for key in dataframes:
    dataframes[key]['year'] = dataframes[key]['din_instante'].dt.year
    dataframes[key]['month'] = dataframes[key]['din_instante'].dt.month
    dataframes[key]['day'] = dataframes[key]['din_instante'].dt.day
    dataframes[key]['hour'] = dataframes[key]['din_instante'].dt.hour
    print(dataframes[key].head(2))
```

```python
for key in dataframes:
    dataframes[key] = dataframes[key].drop(columns=['din_instante', 'day', 'hour'])
    print(dataframes[key].head(2))
```

```python
for key in dataframes:
    dataframes[key] = dataframes[key].groupby(['year', 'month'])['val_geracao'].sum().reset_index()
    print(dataframes[key].head(2))
```

```python
combined_data = pd.concat(dataframes.values(), ignore_index=True)
combined_data
```
